# Series

## creating Series

```python
import pandas as pd # importing pandas
list = [10,2,42,38,27,61]
var = pd.Series(list) # converting list into Seri
print(list)
print(type(list))
print()
print(var)
```

#### Empty Series

```python
empyt = pd.Series([])
empyt
```

#### Changing Index of Series

```python
#method 1
series = pd.Series([11,12,13,14,15], index = ['a','b','c','d','e'])
series
```

```python
#method 2
ind = ['A','B','C','D']
list = [98,90,88,97]
var1 = pd.Series(list, index = ind )
print(var1)
```

#### Naming the Series


```python
var2 = pd.Series([10,20,30,40,50], name = "numbers")
print(var2)
```

#### creating series using dictionary

```python
dist = {'rajeev' : 95,
        'someone' : 93,
        'somebody' : 90}
var3 = pd.Series(dist)
print(dist)
print(type(dist))
print()
print(var3)
print(type(var3))
```

```python
print(var3[0])
```

```python
print(var3[1:])
```

```python
print(max(var3))
```

## DataFrame

```python
data = pd.DataFrame()
print(data)
```

#### Creating dataframe from list

```python
list = [9,8,7,6,5,4,3,2,1]
data = pd.DataFrame(list)
data
```

#### Creating dataframe with multipal col

```python
lst = [[1,2,3,4,5],[10,20,30,40,50],[2,4,6,8,10],[3,6,9,12,15],[1,2,3,5,7]]
data1 = pd.DataFrame(lst)

data1
```

#### Creating dataframe from dictionary

```python
dist = [{'a' : 10, 'b' : 20, 'c' : 30},
       {'a' : 11, 'b' : 12, 'c' : 13}]

data2 = pd.DataFrame(dist)
data2
```

#### creating dataframe from series

```python
import pandas as pd
dist = {'name' : pd.Series(["Rajeev","someone","somebody"]),
        'math' : pd.Series([90,98,88]),
        'phy' : pd.Series([87,78,89])}


dataframe = pd.DataFrame(dist)
dataframe
```
